## Government participation (last 5 proposals)

The validator not only supports infrastructure but keeps the responsibility of network development. This criterion is very easy for understanding: the validator gets 5 points for each vote from the last 5 proposals. 25 points is a maximum.

## Pre-commit (last 100,000 blocks)

The validator should sign blocks permanently. Pre-commits missing might be a signal of the incorrect setup of the validator infrastructure issues.

The calculations are simple: The proportion between validator pre-commits for the last 100,000 blocks and 100,000 blocks (including only participation in one round of the pre-commits) should be multiplied by 15 points. 

## Self-responsibility

This criterium falls into two parts `self delegation` and `min self delegation`.  15 points are proposed for those criteria. 

#### Self delegation

This criterion shows the incentive of the validator to keep his validator node running steadily. In case of slashing the operator will have lost his own tokens. The more tokens are delegated the more incentives. For that criterion 13 points are allocated. 

There are no top limits for self-delegation, so for calculations will be used proportion between self delegated tokens max self delegated tokens in the active validator set. This proportion will be multiplied by the maximum value of this criterion or 13 points.

#### Min self delegation

This is the initial param of the `create validator` transaction. It helps understand possible behavior ways of the validator in the future. The operator can't self-delegate fewer tokens than was set in the current param. Hence, this criterion shows seriousness in the manner of the long-term validating.

Currently, this criterion is controversial,  because few validators really thinking about its importance. To avoid any discussions was decided to allocate only 3 points for that discipline and calculate it by `order-average` scheme, namely:

validators will be sorted by the min self delegation parameter descending, the order places will be distributed (places can be the same), and points will be distributed descending from 3 to 3 divided on validators number (step). In the case of the same order places the average value will get.




## Voting power

This parameter shows the current state of the validator reputation. 10 points are allocated to this criterion. This criterion was estimated to be 10 points because, on the one hand, it is very representative but on another, a lot of delegations were added only by zero-fee ads, moreover a lot of delegations from the foundation. To avoid centralization voting power criterion was moved to the background.

As in the case with self delegation criterion, the proportion between validator voting power and max voting power in the validator set will be used.

## Tokens blurring

The most important criterion. It shows how many tokens the validator lost during the validating cause of slashing. 20 points are allocated.

This criterion is defined as the proportion between validator `tokens` and `delegator shares` params. Normally it should be 1, but in case of slashing it decreases. 

The calculations will be by `order-average` scheme described above.

## Info 

Nice to know something about infrastructure, principles, or position of identity who behind of the validator address. This criterion (5 points) divides into 4 sub-criteria.

#### Identity (2 point)
Most often keybase id

#### website (1 point)
Validator's blog, or side project

#### security contact (1 point)
The source to link validator

####  detail (1 point)
Some kind of validator's position

For all of the named sub-criteria, the binary calculations (if have get points) will be used.


## Commission 

The commission criterion shows more than it looks. Most of the delegators decide to delegate to validators with low or zero commission rates. But sometimes it is no more than ads. Delegators should take a look at the daily commission rate change and at the maximum commission rate for smart delegating. All named criteria can help to understand the validator's possible behavior in the future. This criterion (10 points) divides into 3 sub-criteria. 

#### commission rate (5 point)
The current commission of the validator

#### max commission rate (3 point)
The maximum commission rate validator can set

#### max commission rate change(1 point)
The maximum commission rate validator can change per day

The calculations will be by 'order-average' scheme. In the commission rate, criterion validators with zero commissions will be moved to the end to avoid ads impact.
